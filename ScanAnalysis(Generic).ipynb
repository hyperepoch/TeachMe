## Analyze the data

### Subtask:
Perform necessary calculations and aggregations to prepare the data for visualization based on the requirements for management analysis.


**Reasoning**:
Perform the necessary calculations and aggregations as described in the instructions to prepare the data for visualization.



```python
pd_category_counts = df['PD Category'].value_counts()
display(pd_category_counts)

confidence_averages = df.groupby('PD Category')[['% Content Confidence', '% Header Confidence']].mean()
display(confidence_averages)

unique_owners_count = df['Owners'].nunique()
print(f"Number of unique owners: {unique_owners_count}")

data_type_counts = df['Data Type'].value_counts()
display(data_type_counts)

data_system_type_counts = df['Data System Type'].value_counts()
display(data_system_type_counts)
```

## Visualize the data

### Subtask:
Create appropriate visualizations (e.g., bar charts, line plots, scatter plots) to represent the key findings from the data analysis.


**Reasoning**:
Create the specified bar plots to visualize the data distributions and averages calculated in the previous step.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=pd_category_counts.index, y=pd_category_counts.values)
plt.xlabel('PD Category')
plt.ylabel('Count')
plt.title('Distribution of PD Categories')

plt.figure(figsize=(10, 6))
confidence_averages.plot(kind='bar')
plt.xlabel('PD Category')
plt.ylabel('Average Confidence')
plt.title('Average Content and Header Confidence by PD Category')
plt.xticks(rotation=0)

plt.figure(figsize=(8, 5))
sns.barplot(x=data_type_counts.index, y=data_type_counts.values)
plt.xlabel('Data Type')
plt.ylabel('Count')
plt.title('Distribution of Data Types')

plt.figure(figsize=(8, 5))
sns.barplot(x=data_system_type_counts.index, y=data_system_type_counts.values)
plt.xlabel('Data System Type')
plt.ylabel('Count')
plt.title('Distribution of Data System Types')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols_tables = ['Columns with PD', 'Total Columns', 'Data Element Counts']

plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols_tables):
    plt.subplot(1, 3, i + 1)
    sns.boxplot(y=df_tables[col])
    plt.title(f'Box plot of {col}')

plt.tight_layout()
plt.show()
```

## Analyze the data

### Subtask:
Perform necessary calculations and aggregations to prepare the data for visualization based on the requirements for management analysis.


**Reasoning**:
Perform the necessary calculations and aggregations as described in the instructions to prepare the data for visualization.



```python
total_tables = df_tables.shape[0]
print(f"Total number of tables: {total_tables}")

total_columns = df_tables['Total Columns'].sum()
print(f"Total number of columns: {total_columns}")

total_pd_columns = df_tables['Columns with PD'].sum()
print(f"Total number of columns with PD: {total_pd_columns}")

tables_by_data_system_type = df_tables['Data System Type'].value_counts()
display(tables_by_data_system_type)

tables_by_database_name = df_tables['Database Name'].value_counts()
display(tables_by_database_name)

tables_by_owners = df_tables['Owners'].value_counts()
display(tables_by_owners)

data_element_counts_by_instance = df_tables.groupby('Data System Instance')['Data Element Counts'].sum()
display(data_element_counts_by_instance)

data_element_counts_by_table = df_tables.groupby('Table')['Data Element Counts'].sum()
display(data_element_counts_by_table)

df_tables['% Columns with PD'] = (df_tables['Columns with PD'] / df_tables['Total Columns']) * 100
display(df_tables[['Table', 'Columns with PD', 'Total Columns', '% Columns with PD']].head())

high_pd_tables = df_tables[df_tables['% Columns with PD'] > 50]
display(high_pd_tables[['Table', '% Columns with PD']])
```

## Visualize the data

### Subtask:
Create appropriate visualizations (e.g., bar charts, line plots, scatter plots) to represent the key findings from the data analysis.


**Reasoning**:
Create the specified visualizations to represent the key findings from the data analysis.



```python
plt.figure(figsize=(10, 6))
sns.barplot(x=tables_by_data_system_type.index, y=tables_by_data_system_type.values)
plt.xlabel('Data System Type')
plt.ylabel('Number of Tables')
plt.title('Distribution of Tables by Data System Type')

plt.figure(figsize=(10, 6))
sns.barplot(x=tables_by_database_name.index, y=tables_by_database_name.values)
plt.xlabel('Database Name')
plt.ylabel('Number of Tables')
plt.title('Distribution of Tables by Database Name')

plt.figure(figsize=(10, 6))
sns.barplot(x=tables_by_owners.index, y=tables_by_owners.values)
plt.xlabel('Owner')
plt.ylabel('Number of Tables')
plt.title('Distribution of Tables by Owners')
plt.xticks(rotation=45, ha='right')

plt.figure(figsize=(12, 6))
sns.barplot(x=data_element_counts_by_instance.index, y=data_element_counts_by_instance.values)
plt.xlabel('Data System Instance')
plt.ylabel('Total Data Element Counts')
plt.title('Total Data Element Counts by Data System Instance')
plt.xticks(rotation=45, ha='right')

plt.figure(figsize=(15, 7))
top_n = 10
sns.barplot(x=data_element_counts_by_table.sort_values(ascending=False).head(top_n).index, y=data_element_counts_by_table.sort_values(ascending=False).head(top_n).values)
plt.xlabel('Table')
plt.ylabel('Total Data Element Counts')
plt.title(f'Top {top_n} Tables by Total Data Element Counts')
plt.xticks(rotation=45, ha='right')

plt.figure(figsize=(8, 6))
sns.boxplot(y=df_tables['% Columns with PD'])
plt.ylabel('% Columns with PD')
plt.title('Distribution of % Columns with PD per Table')

plt.tight_layout()
plt.show()
```
